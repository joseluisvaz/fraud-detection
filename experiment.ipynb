```python
# import all dependencies and data
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
import seaborn as sns
import calendar
from sklearn import neural_network

import utils

%matplotlib notebook
pd.set_option('chained_assignment',None)
sns.set_style("white")

data = pd.read_csv('claims.csv')
#cleaning temporal data
data = utils.clean_data(data)
```

```python
counted = data.groupby(["FraudFound_P"]).size()
```

```python
#features to categorize
features = ["Make","WeekOfMonth", "WeekOfMonthClaimed", "Month", "DayOfWeek", 
            "AccidentArea", "DayOfWeekClaimed", "MonthClaimed", "Sex", "MaritalStatus", 
            "Age", "Fault", "PolicyType", "VehicleCategory", "VehiclePrice", 
            "DriverRating", "PastNumberOfClaims", 
            "AgeOfVehicle", "AgeOfPolicyHolder", "PoliceReportFiled", "WitnessPresent", 
            "AgentType", "NumberOfSuppliments", "AddressChange_Claim", "NumberOfCars", 
            "Year", "BasePolicy"]
```

```python
Ydata = data["FraudFound_P"]
data = data[features]
for i in features:
    data[i] = data[i].astype('category')
```

```python
#Dummies and to numpy array
Xdata = pd.get_dummies(data).as_matrix()
Ydata = Ydata.as_matrix()
```

```python
clf2 = neural_network.MLPClassifier(hidden_layer_sizes= (50,20,4))
```

```python
clf2.fit(Xdata, Ydata)
```

```python
np.sum(Ydata != clf2.predict(Xdata))
```

```python
ydatadf = pd.DataFrame(Ydata)
```

```python
condition = pd.DataFrame(Ydata != clf2.predict(Xdata))
```

```python
final  = pd.concat([ydatadf, condition], axis =  1)
final.columns = ['Ydata', 'condition']
```
